# Part 2. Text Analysis

## 2.1 extract high requency words in description

```python
import re
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.probability import *
from itertools import chain
#from tqdm import tqdm
import codecs
```

```python
with open('stopwords.txt') as f:
    stop_words = f.read().splitlines()
stop_words = set(stop_words)
print(stop_words)
```

```python
# write your code here
# define your tokenizer
tokenizer = RegexpTokenizer(r"\w+(?:[-']\w+)?") 
filtered_description = []
# index = 0
# print(stop_words)
for descp in wine_data['description']:
#     index += 1
    tokenized_descp = tokenizer.tokenize(descp)
#     print(tokenized_descp)
    for w in tokenized_descp: 
        if w.casefold() not in stop_words: 
            filtered_description.append(w)
#     print(filtered_description)
#     if index == 10:
#         break
```

```python
# find top common words with document frequencies > 5000
# you may use function FreqDist() and sort()
freq_descp = FreqDist(filtered_description)
freq_descp_df = pd.DataFrame(list(freq_descp.items()), columns = ["Word","Frequency"])
most_common_descprition = freq_descp_df[freq_descp_df['Frequency']>5000]
sorted_common_description = most_common_descprition.sort_values("Frequency", ascending = False)
sorted_common_description
```

```python
# save your table to 'top_common_words.txt'  
sorted_common_description.to_csv("top_common_words.txt")
```

## 2.2 Find key words for describing Shiraz using TF-IDF

```python
# select 'description' from 'variety' eqaul to  'Shiraz' 
corpus = wine_data[wine_data['variety'] == 'Shiraz']['description']
```

```python
# use TfidfVectorizer to calculate TF-IDF score
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(analyzer='word', stop_words = 'english')
#tf is term frequency... for certain word, it is: total occurance of that word in a document/total words in that document
#idf is inverse document frequency, it is: log of (total no. of documents/no. of document containing that word)
#tf-idf for a word is tf*idf calculated.

tfidf_matrix =  tfidf.fit_transform(corpus)
# print(tfidf_matrix)
#output is like (A,B), C
#A: Document index, B: Specific word-vector index, C: TFIDF score for word B in document A

# print(tfidf_matrix.shape)
#In this matrix, row is the number of documnets and column is the number of unique words.
feature_names = tfidf.get_feature_names()
# print(feature_names)

doc = 821
feature_index = tfidf_matrix[doc,:].nonzero()[1]
tfidf_scores = zip(feature_index, [tfidf_matrix[doc, x] for x in feature_index])

description_dframe = pd.DataFrame(columns = ['feature_name', 'tfidf_score'])
for w, s in [(feature_names[i], s) for (i, s) in tfidf_scores]:
        description_dframe = description_dframe.append({'feature_name': w, 'tfidf_score': s}, ignore_index=True)
print(description_dframe)
```
